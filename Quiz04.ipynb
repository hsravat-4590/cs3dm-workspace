````python
import pandas as pd

X = pd.read_csv("tcruzi.csv")
X
````

````python
X.drop(columns=X.columns[0], axis=1, inplace=True)
X
````

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=[0,1])
X = scaler.fit_transform(X)
X
````

````python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(affinity='euclidean', linkage='ward')

y_hc = hc.fit_predict(X)
````

````python
# visualise:
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = [10, 10]
plt.scatter(X[:,0], X[:,1], c=y_hc, cmap='viridis')
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
````

````python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2, init='random',
                  learning_rate='auto', perplexity=5,
                  random_state=1234).fit_transform( X )
````

````python
plt.scatter(X_embedded[:,0], X_embedded[:,1], c=y_hc, cmap='viridis')
plt.show()
````
